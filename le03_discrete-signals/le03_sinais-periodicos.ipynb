## Desafio: Gerando sinais periódicos

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The sinusoid

n = np.arange(100)
fs = 100 # frequencia de amostragem
f = 3 # frequencia do sinal

w0 = 2 * np.pi * f / fs # frequencia digital
theta = 90 * np.pi / 180 # fase
A = 220 # amplitude

n = np.arange(100)
x = A * np.sin(w0 * n + theta)

fig, ax = plt.subplots()
ml, sl, bl = plt.stem(n, x, linefmt='r', markerfmt='or', basefmt='lime')
plt.setp(sl, 'linewidth', .5)
plt.setp(ml, 'markersize', 5)
plt.setp(bl, 'linewidth', .5)
```

```python
# AWGN

P = 50

g = P * np.random.randn(len(n))

fig, ax = plt.subplots()
ml, sl, bl = plt.stem(n, g, linefmt='b', markerfmt='ob', basefmt='lime')
plt.setp(sl, 'linewidth', .5)
plt.setp(ml, 'markersize', 5)
plt.setp(bl, 'linewidth', .5)
```

```python
# Sinal resultado

y = x + g

fig, ax = plt.subplots()
ml, sl, bl = plt.stem(n, y, linefmt='g', markerfmt='og', basefmt='lime')
plt.setp(sl, 'linewidth', .5)
plt.setp(ml, 'markersize', 5)
plt.setp(bl, 'linewidth', .5)
```
